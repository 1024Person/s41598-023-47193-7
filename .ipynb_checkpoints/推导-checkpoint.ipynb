# 推导$F(s)$

```python
import sympy as sp
```

```python
g,l,b,lb,w0,t,tp = sp.symbols(r'gamma,lambda,beta,lambda_1,omega_0,t,t_1')
s =sp.symbols('s')
F = g*l/4*sp.cosh(b*(lb+sp.I*w0)*t)*sp.exp(-lb*t)
F
```

```python
r = sp.laplace_transform(F,t,s)[0]
r
```

## 解决办法
提前将想要的数据带入进去，然后这个拉普拉斯逆变换就容易变换出来了，最后画图就行了,但是这样也有一个缺点就是这样的话，就没法知道完全的解析解了，所以只能用数值的。但是看论文的话，好像也没有太多的影响。

```python
# ======================================== γ = 0.1 ================================
# ======================================== β = 0 ==================================
D = sp.Symbol('D')
Delta = sp.Symbol("Delta")
r1p_b0 = 1/(s+r+sp.I*D)
r2p_b0 = r1p_b0.subs({w0:1.5e9,D:0.3,lb:l-sp.I*Delta,b:0,g:0.1})
r3p_b0 = sp.simplify(r2p_b0.subs({Delta:0,l:1}))
Mp_b0 = sp.simplify(sp.inverse_laplace_transform(r3p_b0,s,t))

r1m_b0 = 1/(s+r-sp.I*D)
r2m_b0 = r1m_b0.subs({w0:1.5e9,D:0.3,lb:l-sp.I*Delta,b:0,g:0.1})
r3m_b0 = sp.simplify(r2m_b0.subs({Delta:0,l:1}))
Mm_b0 = sp.simplify(sp.inverse_laplace_transform(r3m_b0,s,t))
r3p_b0
```

```python
#========================================β=3e-9 ====================================
r1p_b3 = 1/(s+r+sp.I*D)
r2p_b3 = r1p_b3.subs({w0:1.5e9,D:0.3,lb:l-sp.I*Delta,b:3e-9,g:0.1})
r3p_b3 = sp.simplify(r2p_b3.subs({Delta:0,l:1}))
Mp_b3 = sp.inverse_laplace_transform(r3p_b3,s,t)
r1m_b3 = 1/(s+r-sp.I*D)
r2m_b3 = r1m_b3.subs({w0:1.5e9,D:0.3,lb:l-sp.I*Delta,b:3e-9,g:0.1})
r3m_b3 = sp.simplify(r2m_b3.subs({Delta:0,l:1}))
Mm_b3 = sp.simplify(sp.inverse_laplace_transform(r3m_b3,s,t))
```

```python
# #====================================== β=8e-9 =====================================
r1p_b8 = 1/(s+r+sp.I*D)
r2p_b8 = r1p_b8.subs({w0:1.5e9,D:0.3,lb:l-sp.I*Delta,b:8e-9,g:0.1})
r3p_b8 = sp.simplify(r2p_b8.subs({Delta:0,l:1}))
Mp_b8 = sp.simplify(sp.inverse_laplace_transform(r3p_b8,s,t))
r1m_b8 = 1/(s+r-sp.I*D)
r2m_b8 = r1m_b8.subs({w0:1.5e9,D:0.3,lb:l-sp.I*Delta,b:8e-9,g:0.1})
r3m_b8 = sp.simplify(r2m_b8.subs({Delta:0,l:1}))
Mm_b8 = sp.simplify(sp.inverse_laplace_transform(r3m_b8,s,t))
```

```python
#========================================β=3e-9 ====================================
r1p_b3 = 1/(s+r+sp.I*D)
r2p_b3 = r1p_b3.subs({w0:1.5e9,D:0.3,lb:l-sp.I*Delta,g:0.1})# 大不了就先不替换β了呗，等都算出来之后我们再把β给替换掉
# python的这个库用来计算难算的是真的不行！！！
r3p_b3 = sp.simplify(r2p_b3.subs({Delta:0,l:1}))
# r3p_b3
Mp_b3 = sp.inverse_laplace_transform(r3p_b3,s,t)
Mp_b3
```
